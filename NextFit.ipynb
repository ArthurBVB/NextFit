### Identifique o per√≠odo do dia (manh√£, tarde e noite) que em m√©dia recebe mais check_ins.

primeiramente vamos verificar o valor que mais aparece, ele fica como resultado chamado Top no comando describe

```python
base.tempo_medio_chegada.describe()
```

porem a pergunta seria o periodo que mais faz check-in manha tarde ou noite logo precisamos categorizar sendo:
 manha das 00:01 ate as 12:00
 tarde das 12:01 ate 18:00
 noite das 18:01 ate 00:00

```python
# vamos fazer uma logica para retornar manha tarde ou noite com if else, como √© um objeto precisa colocar aspas
def turno(tempo_medio_chegada):
  if '00:00:00' <= tempo_medio_chegada < '12:00:00':
    return 'manh√£'
  elif '12:00:00' <= tempo_medio_chegada < '18:00:00':
    return 'tarde'
  else:
    return 'noite'
```

```python
# aplicando a formula e trazendo como uma nova coluna
base['periodo_checkin'] = base['tempo_medio_chegada'].apply(turno)
```

```python
# verificando se retornou todos valores e nenhum deles nulo, tambem fiz um base.head() para verificar visualmente
base.info()
```

```python
base.periodo_checkin.value_counts()
```

```python
sns.countplot(x=base.periodo_checkin, palette="pastel")
plt.title('Contagem por Periodo')
plt.xlabel('Periodo')
plt.ylabel('Contagem')
plt.show()
```
